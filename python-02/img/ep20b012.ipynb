```python
def conway(init: np.ndarray, niter:int)->np.ndarray:
    if niter !=0:
        a = init.copy()
        n = np.shape(a)[0]
        new_array = np.zeros(np.shape(a),dtype=int)
        for i in range(1,n-1):
            for j in range(1,n-1):
                ans = np.sum(a[i-1:i+2,j-1:j+2])
                if ans==3:
                    new_array[i,j] = 1
                elif ans==4 and a[i,j]==1:
                    new_array[i,j] = 1
                else:
                    new_array[i,j]=0
        return conway(new_array,niter-1)
    elif niter==0:
        return init
```
